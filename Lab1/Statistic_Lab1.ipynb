## Ход выполнения работы
$\Omega$ - пространство элементарных исходов, заданное как квадрат со стороной $2a$. $\mathcal{A} (r)$ - благоприятное событие, круг радиуса $r$ с центром в начале координат.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

В качестве значений радиуса круга мы выбрали значения $r = \frac{a}{k + 1}$, где $k = 0, 1, 2, 3, 4$. Для каждого значения $r$ построили круг и нашли геометрическую вероятность события, заданного окружностью радиуса $r$. Построили графики геометрической вероятности события для различных радиусов кругов.

```python
a = 10  # половина стороны квадрата элементарных исходов
r = [round(a / (k + 1), 1) for k in range(5)]  # радиусы кругов

for r_c in r:
    plt.figure(figsize=(5, 5))
    plt.title(f'Событие заданное окружностью радиуса {r_c}')
    plt.xlim(-a, a)
    plt.ylim(-a, a)
    circle = plt.Circle((0, 0), r_c, color='r', fill=True)
    plt.gca().add_artist(circle)
    plt.show()
```

Реализовали функцию для вычисления геометрической вероятности события для различных радиусов кругов. Для этого потребовалось воспользоваться формулой $p(\mathcal{A}) = \frac{S(A)}{S(\Omega)} = \frac{\pi r^2}{(2a)^2}$. Далее мы посчитали истинную геометрическую вероятность события для различных радиусов кругов с помощью данной функции.

```python
def geometric_probability(r, a):
    return (np.pi * r ** 2  # площадь круга
            / ((2 * a) ** 2))  # площадь квадрата


print('Геометрическая вероятность события для различных радиусов кругов:')
for r_c in r:
    print(f'радиуса {r_c} - {geometric_probability(r_c, a)}')
```

Из полученных результатов можно сделать вывод, что при увеличении радиуса круга, геометрическая вероятность события увеличивается. Это связано с тем, что при увеличении радиуса круга увеличивается площадь круга, а значит увеличивается вероятность попадания точки в круг.

Далее при помощи генератора случайных чисел мы сгенерировали 5 случайных точек в пределах квадрата и округлили координаты точек до 1 знака после запятой. Реализовали функцию для проверки принадлежности точки событию, проверили принадлежность точек событию для различных радиусов кругов и построили графики, отображающие принадлежность точек событию. Для генерации чисел было использовано равномерное распределение.

```python
points = np.random.uniform(-a, a, (5, 2))  # 5 случайных точек в пределах квадрата
points = np.round(points, 1)  # округление координат точек до 1 знака после запятой


def in_circle(r, point):
    return point[0] ** 2 + point[1] ** 2 <= r ** 2  # проверка попадания точки в круг с помощью уравнения окружности


print('Проверка принадлежности точки событию:')
for r_c in r:
    plt.figure(figsize=(7, 7))
    plt.title(f'Событие заданное окружностью радиуса {r_c}')
    plt.xlim(-a, a)
    plt.ylim(-a, a)
    circle = plt.Circle((0, 0), r_c, color='r', fill=False)
    plt.gca().add_artist(circle)
    for point in points:
        plt.scatter(point[0], point[1], color='b', s=40)
        plt.text(point[0], point[1] + 0.5, f'{point}', fontsize=12)
    plt.show()

    for point in points:
        if in_circle(r_c, point):
            print(f'Точка {point} принадлежит событию')
        else:
            print(f'Точка {point} не принадлежит событию')
    print()
```

Далее реализовали функцию для подсчёта доли точек, попавших в круг, и построили графики зависимости доли точек, попавших в круг, от количества точек. Для этого были использованы значения $n = 1, 2, ..., 1000$. Для каждого значения $n$ было сгенерировано $n$ случайных точек в пределах квадрата, после чего была подсчитана доля точек, попавших в круг.

```python
def points_in_circle_ratio(n, r, a):  # доля точек из n, попавших в круг
    points = np.random.uniform(-a, a, (n, 2))  # n случайных точек в пределах квадрата
    in_circle_points = 0  # количество точек, попавших в круг
    for point in points:
        if in_circle(r, point):  # проверка попадания точки в круг
            in_circle_points += 1
    return in_circle_points / n  # доля точек, попавших в круг


for r_c in r:
    ns = range(1, 1000)
    p = [points_in_circle_ratio(n, r_c, a) for n in ns]

    plt.figure(figsize=(10, 5))
    plt.scatter(ns, p, label='ˆp(n)')
    plt.xlabel('Количество точек')
    plt.ylabel('Доля точек, попавших в круг')
    plt.title(f'График ˆp(n) для радиуса круга r = {r_c}')
    plt.show()
```

Анализируя графики, можно заметить, что при увеличении количества точек, попавших в круг, доля точек, попавших в круг, стремится к истинной геометрической вероятности события. Таким образом точность оценки геометрической вероятности события увеличивается с увеличением количества точек.

Далее реализовали функцию для анализа изменения ошибки оценки геометрической вероятности события и построили графики зависимости ошибки оценки геометрической вероятности события от количества точек.

```python
def error_in_circle_ratio(n, r, a):
    return np.abs(points_in_circle_ratio(n, r, a) - geometric_probability(r, a))


for r_c in r:
    ns = np.arange(1, 1000)
    e = [error_in_circle_ratio(n, r_c, a) for n in ns]

    plt.figure(figsize=(10, 5))
    plt.scatter(ns, e)
    plt.xlabel('Количество точек')
    plt.ylabel('Ошибка оценки геометрической вероятности')
    plt.title(f'График $\\epsilon$(n) для радиуса круга r = {r_c}')
    plt.show()
```

Из графиков видно, что при увеличении количества точек ошибка оценки геометрической вероятности события уменьшается.

После того как мы зафиксировали последовательность $\{\varepsilon\}$, для каждого значения $r$ нашли такое количество случайных точек $N$, что ошибка оценки геометрической вероятности события не превышает $\varepsilon$. Для каждого значения $r$ построили графики зависимости $N(\varepsilon)$.

```python
eps = [0.01, 0.001, 0.0001, 0.00001]
eps_string = [f'{e}' for e in eps]


def find_n_for_eps(eps, s, r, a): # нахождение N для заданной точности и шага
    N = s
    while error_in_circle_ratio(N, r, a) > eps:
        N += s
    return N


for r_c in r:
    ns = [find_n_for_eps(e, 100, r_c, a) for e in eps]

    plt.figure(figsize=(10, 5))
    plt.plot(eps_string, ns)
    for point in zip(eps_string, ns):
        plt.text(point[0], point[1], f'{point[1]}', fontsize=10)
    plt.xlabel('Точность')
    plt.ylabel('Количество точек')
    plt.title(f'График N($\\epsilon$) для радиуса круга r = {r_c}')
    plt.show()
```

Анализируя графики, можно сделать вывод, что при уменьшении радиуса круга увеличивается количество точек, необходимых для оценки геометрической вероятности события с заданной точностью. Кроме того, при увеличении точности увеличивается количество точек, необходимых для оценки геометрической вероятности события.

Стоит обратить внимание на график 3. Количество точек, необходимых для достижения последнего значения $\varepsilon$, не увеличилось по сравнению с предыдущим значением $\varepsilon$. Можно предположить, что это произошло из-за рандомного характера генерации точек. Скорее всего, при другой генерации точек количество точек для достижения $\varepsilon$ увеличилось бы, как и на остальных графиках.

Далее реализовали функцию для усреднения значений $N$ для каждого значения $\varepsilon$ и построили графики зависимости усреднённого $N$ от $\varepsilon$. Это позволило нам оценить среднее количество точек, необходимых для оценки геометрической вероятности события с заданной точностью и более точно определить зависимость количества точек от точности.

```python
for r_c in r:
    ns = [sum([find_n_for_eps(e, 100, r_c, a) for _ in range(10)]) / 10 for e in eps]

    plt.figure(figsize=(10, 5))
    plt.plot(eps_string, ns)
    for point in zip(eps_string, ns):
        plt.text(point[0], point[1], f'{point[1]}', fontsize=10)
    plt.xlabel('Точность')
    plt.ylabel('Количество точек')
    plt.title(f'График усреднённого N($\\epsilon$) для 10 испытаний и для радиуса круга r = {r_c}')
    plt.show()
```

Благодаря усреднению значений $N$ для каждого значения $\varepsilon$ мы смогли получить более точные результаты и избавиться от случайных колебаний. На этих графиках лучше видно, что при уменьшении радиуса круга увеличивается количество точек, необходимых для оценки геометрической вероятности события с заданной точностью.
